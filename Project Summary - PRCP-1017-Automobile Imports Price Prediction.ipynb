## Project Summary

Determining the price of a car is a challenging task, due to the many factors that drive a vehicleâ€™s price on the market. The focus of this project is to develope machine learning models that can accurately predict the price of a car based on its features. Mainly, there are two project objectives:

- To find the variables which are significant in predicting the car price
- To find most relevant machine learning model which predicts best car price for given dataset

The input file is given as an excel csv file which contains 201 records of different cars information and 26 total features which are very well defined. The shape of the data is (201 X 26). These 26 features are classified as 16 numerical and 10 categorical data. The dataset has total 51 missing values which are denoted as "?". Based on the Normality tests of numerical variables with missing values, it has been decided to replace all numerical missing values with Median values of that perticular column. Categorical missing values are replaced with Mode value of that column.

The Machine Learning Regressor Models are used to find the R2-score & Mean_absolute_error of the data. Two types of datasets have been used : One dataset consists of All Outliers of Predictor and Target Variables where as in second dataset Outliers of Target variables are dropped. The models used are:

1. Linear Regression
2. XG Boost Regression
3. Random Forest Regression

From these models XG Boost Regressor algorithm gives the better R2-Score of 95% and lesser Mean_absolute_error of 992.

### Machine Learning Model 
The machine learning models used in this project are

- 1.Linear Regression 

    Linear Regression was chosen as the first model due to its simplicity and comparatively small training time.
    
    
- 2.RandomForestRegressor

    Random Forest is an ensemble learning based regression model. It uses a model called decision tree, specifically as the name suggests, multiple decision trees to generate the ensemble model which collectively produces a prediction. The benefit of this model is that the trees are produced in parallel and are relatively uncorrelated, thus producing good results as each tree is not prone to individual errors of other trees. 
    
    
- 3.XGBoost Regressor
    
    Extreme Gradient Boosting or XGBoost is one of the most popular machine learning models in current times. XGBoost is quite similar at the core to the original gradient boosting algorithm but features many additive features that significantly improve its performance such as built in support for regularization, parallel processing as well as giving additional hyperparameters to tune such as tree pruning, sub sampling and number of decision trees. A maximum depth of 5 was used and the algorithm was run on all cores in parallel

All these machine learning algorithms are best for regression problem. The train and test data are divided into 80:20 ratio and fitted into the model and passed through the machine learning. The dataset here used is including All outliers of Predictor and Target variables.

The predicted data and test data achieved the accuracy rate of, 

* 1.Linear Regression - 80% with Mean_absolute_error of 2250
* 2.RandomForestRegressor - 88.5% with Mean_absolute_error 1658
* 3.XGBoost Regressor - 95.1% with Mean_absolute_error 992
